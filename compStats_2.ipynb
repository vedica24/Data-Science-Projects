```python
import numpy as np
from scipy.stats import bernoulli
from scipy import optimize
import matplotlib.pyplot as plt
```

Question 1) - Random Walk on integers with probability of moving right as p, probability of moving left 1-p
- Initial state distribution: P(Xo=0) = 1

```python
for p in [0.5, 1/3, 2/3, 0.99, 0.01]:
  arr = []
  for j in range(1000):
    x=0
    for i in np.random.binomial(size=1000, n=1, p= p):
      if i == 0:
        x = x-1
      else:
        x = x+1
    arr.append(x)
  n, bins, ignored = plt.hist(arr, 25, color ='orange', alpha = 0.6, edgecolor = "black")
  plt.title(p)
  plt.show()
  plt.clf()
  print()
  print()
  print()
```

Question 2) Initial state distribution is discrete in the case given below

```python
U = np.random.uniform(0,1,1000)
arr = []
for i in U:
  if i<=0.3:
    arr.append(-1)
  elif i<=0.4:
    arr.append(0)
  elif i<=0.5:
    arr.append(1)
  elif i<=0.9:
    arr.append(2)
  else:
    arr.append(3)

arr2 = arr
n, bins, ignored = plt.hist(arr2, 25, color ='orange', alpha = 0.6, edgecolor = "black")
plt.title('arr')
print()
plt.show()
for p in [0.5, 1/3, 2/3, 0.99, 0.01]:
  arr3 = []
  for x in arr2:
    for i in np.random.binomial(size=1000, n=1, p= p):
      if i ==0:
        x= x-1
      else:
        x= x+1
    arr3.append(x)
  n, bins, ignored = plt.hist(arr3, 25, color ='orange', alpha = 0.6, edgecolor = "black")
  plt.title(p)
  plt.show()
  plt.clf()
  print()
  print()
```

Question 3) Random Walk on 2D
- Initial state P(Xo=(0,0)) = 1

```python
n = 100
p1 = 0.5
p2 = 0.5

r1 = bernoulli.rvs(p1, size = n)

r2 = bernoulli.rvs(p2, size = n)

X = np.zeros(n+1)
Y = np.zeros(n+1)

XY = np.zeros((n+1,2))

for t in range(n):
  if r1[t] == 1:
    X[t+1] = X[t] + 1
    XY[t+1,0] = XY[t,0] + 1
  else:
    X[t+1] = X[t] - 1
    XY[t+1,0] = XY[t,0] - 1
  if r2[t] == 1:
    Y[t+1] = Y[t] + 1
    XY[t+1,1] = XY[t,1] + 1
  else:
    Y[t+1] = Y[t] - 1
    XY[t+1,1] = XY[t,1] - 1

  plt.scatter(XY[t+1,0], XY[t+1,1],color = 'orange', alpha = 0.6)
  ax = plt.subplot(111)
  ax.annotate("", xy = (XY[t+1]), xycoords = 'data', xytext = (XY[t]), textcoords = 'data', arrowprops = dict(arrowstyle = 'simple', connectionstyle = 'arc3'))
```
