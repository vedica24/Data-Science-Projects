```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("training_diabetes.csv")
```

```python
df_test = np.loadtxt("test_diabetes.csv", delimiter = ",")
```

```python
df.describe()
```

#### no missing values, can replace them by mean, mode value, or delete the rows

```python
df
```

```python
X_train = df.loc[:, df.columns != "Y"]
```

```python
Y_train = df.loc[:, "Y"]
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X_train, Y_train)
```

```python
y_predict = reg.predict(df_test)
```

```python
y_train_predict = reg.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
mse_self_implement = mean_squared_error(Y_train, y_train_predict)
 
print("MSE: ", mse_self_implement)

u = ((Y_train - y_train_predict)**2).sum()
v = ((Y_train - Y_train.mean())**2).sum()
score = 1 - u/v
print("Train score : ", score)
```

```python
#weights and intercept term
weights = reg.coef_          
intercept = reg.intercept_
print("Coeffecient for x is:", weights)
print("Intercept value is:", intercept) 
```
